```python
import pandas as pd
import plotly.express as px
```

```python
dados = pd.read_csv('telecom_users.csv')
dados.head(5)
```

```python
dados = dados.drop('Unnamed: 0', axis=1) # Removendo a coluna 'Unnamed: 0'
dados
```

```python
dados.info()
```

```python
dados.TotalGasto = pd.to_numeric(dados.TotalGasto, errors = 'coerce') # transformando valores que são numeros mas o python entendeu como texto
dados.info()
```

```python
dados = dados.dropna(how = 'all', axis = 1) # removendo colunas completamente vazias
dados = dados.dropna(how = 'any', axis = 0) # removendo linha com pelo menos 1 valor vazio
dados.info()
```

```python
dados.Churn.value_counts()
```

```python
dados.Churn.value_counts(normalize = True).map('{:.1%}'.format)
```

```python
for coluna in dados.columns:
    grafico = px.histogram(dados, x = coluna, color = 'Churn')
    grafico.show()
```
